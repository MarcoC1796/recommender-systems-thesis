```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Define the function
def f(q, p):
    r = 5
    # q = theta[0]
    # p = theta[1]
    return (
        np.square(q**2 - 1000) / 2
        + np.square(p**2 - 1000) / 2
        + np.square(p * q - 1000) / 2
    )


# Function to plot 3D function on a given axis
def plot_3d_function(f, ax, center=(0, 0), bounds=40):
    # Generate data
    x = np.linspace(center[0] - bounds, center[0] + bounds, 1000)
    y = np.linspace(center[1] - bounds, center[1] + bounds, 1000)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)

    # Plot the function
    ax.plot_surface(X, Y, Z, cmap="terrain")

    # Customize the plot
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title("3D Function Plot")


# Function to plot contour of a 2D function on a given axis
def plot_contour(ax, f):
    # Generate data
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)

    # Plot the contour with 'terrain' colormap
    contour = ax.contourf(X, Y, Z, cmap="terrain")

    # Add a colorbar
    plt.colorbar(contour, ax=ax)

    # Customize the plot
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Contour Plot")


# Create subplots
fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(231, projection="3d")
ax2 = fig.add_subplot(232, projection="3d")
ax3 = fig.add_subplot(233, projection="3d")
ax4 = fig.add_subplot(234)

# Plot the 3D function on each axis
plot_3d_function(f, ax1)
plot_3d_function(f, ax2, bounds=50)
plot_3d_function(f, ax3, bounds=75)
plot_contour(ax4, f)

# Adjust subplot spacing
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Function to plot 3D function on a given axis
def plot_3d_function(ax, func, center=(0, 0), bounds=40):
    # Generate data
    x = np.linspace(center[0] - bounds, center[0] + bounds, 1000)
    y = np.linspace(center[1] - bounds, center[1] + bounds, 1000)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    # Plot the function
    ax.plot_surface(X, Y, Z, cmap="terrain")

    # Customize the plot
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title("3D Function Plot")


# Function to plot contour of a 2D function on a given axis
def plot_contour(ax, func, center=(0, 0), bounds=40):
    # Generate data
    x = np.linspace(center[0] - bounds, center[0] + bounds, 1000)
    y = np.linspace(center[1] - bounds, center[1] + bounds, 1000)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    # Plot the contour with 'terrain' colormap
    contour = ax.contourf(X, Y, Z, cmap="terrain")

    # Add a colorbar
    plt.colorbar(contour, ax=ax)

    # Customize the plot
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Contour Plot")


# Define the functions to plot
def function1(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


def function2(x, y):
    return np.cos(x) + np.sin(y)


def function3(x, y):
    return np.exp(-np.sqrt(x**2 + y**2))


def function4(x, y):
    return np.sin(x) * np.cos(y)


functions = [function1, function2, function3, function4]


# Determine the number of functions
num_functions = len(functions)

# Calculate the number of rows and columns for the subplot grid
num_rows = num_functions
num_cols = 2

# Create the subplot grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 4 * num_rows))

for ax in axes.flatten():
    ax.axis("off")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(top=False, bottom=False, left=False, right=False)

# Iterate over the functions and plot them in the subplot grid
for i, func in enumerate(functions):
    # Plot the 3D function in the first column
    ax_3d = fig.add_subplot(num_rows, num_cols, i * num_cols + 2, projection="3d")
    plot_3d_function(ax_3d, func, bounds=5)

    # Plot the contour in the second column
    ax_contour = fig.add_subplot(num_rows, num_cols, i * num_cols + 1, frame_on=False)

    plot_contour(ax_contour, func, bounds=5)


# Adjust subplot spacing
plt.subplots_adjust(wspace=0.25, hspace=0.4)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Define the function
def f(q, p):
    r = 5

    return (
        np.square(q**2 - 10) / 2
        + np.square(p**2 - 10) / 2
        + np.square(p * q - 10) / 2
    )


# Function to plot 3D function on a given axis
def plot_3d_function(ax, func, center=(0, 0), bounds=40):
    # Generate data
    x = np.linspace(center[0] - bounds, center[0] + bounds, 1000)
    y = np.linspace(center[1] - bounds, center[1] + bounds, 1000)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    # Plot the function
    ax.plot_surface(X, Y, Z, cmap="coolwarm")

    # Customize the plot
    ax.set_xlabel("q")
    ax.set_ylabel("p")
    ax.set_zlabel("J")
    ax.set_title("3D Function Plot")


# Function to plot contour of a 2D function on a given axis
def plot_contour(ax, func, center=(0, 0), bounds=40):
    # Generate data
    x = np.linspace(center[0] - bounds, center[0] + bounds, 1000)
    y = np.linspace(center[1] - bounds, center[1] + bounds, 1000)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    # Plot the contour with 'terrain' colormap
    contour = ax.contourf(X, Y, Z, cmap="coolwarm")

    # Add a colorbar
    plt.colorbar(contour, ax=ax)

    # Customize the plot
    ax.set_xlabel("q")
    ax.set_ylabel("p")
    ax.set_title("Contour Plot")


functions = [f] * 3


# Determine the number of functions
num_functions = len(functions)

# Calculate the number of rows and columns for the subplot grid
num_rows = num_functions
num_cols = 2

# Create the subplot grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6 * num_rows))

for ax in axes.flatten():
    ax.axis("off")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(top=False, bottom=False, left=False, right=False)

bounds = 0.5
j = 1
# Iterate over the functions and plot them in the subplot grid
for i, func in enumerate(functions):
    xpos = (i % num_cols) * 0.5
    ypos = (num_rows - i - 1) * 0.33
    # Plot the 3D function in the first column
    ax_3d = fig.add_subplot(num_rows, num_cols, i * num_cols + 2, projection="3d")
    plot_3d_function(ax_3d, func, bounds=bounds)

    # Plot the contour in the second column
    ax_contour = fig.add_subplot(num_rows, num_cols, i * num_cols + 1)
    plot_contour(ax_contour, func, bounds=bounds)

    bounds += 1.8

# Adjust subplot spacing
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Show the plot
plt.show()
```
