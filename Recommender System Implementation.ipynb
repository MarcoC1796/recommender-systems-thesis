```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

https://eigentaste.berkeley.edu/dataset/

```python
jokes_df = pd.read_excel('jokes_dataset.xls', header=None, usecols=range(1,151))  
```

```python
jokes_df.shape
```

```python
jokes_df.head()
```

```python
jokes_df = jokes_df.astype(float)
jokes_df.dtypes
```

```python
jokes_df = jokes_df.replace(99, np.nan)
jokes_df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(jokes_df.notna(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.displot(
    data=jokes_df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
plt.show()
```

```python
R = jokes_df.to_numpy()
```

```python
print(f"{round((np.count_nonzero(~np.isnan(R))/R.size)*100,2)}% of full entries")
```
